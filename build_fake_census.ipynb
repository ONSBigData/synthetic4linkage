## Date of Birth 
generate datetime object first then derive variables 
should be between begining of 1904 to end of 2019
age should be integer - age given as age on final day of 2019 
full_DOB should be string DDMMYYY

```python
def calculate_age_on_31_12_2019(born):
    today = datetime.date(2019, 12, 31)
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
def generate_date_of_birth(df):
    day_list = []
    month_list = []
    year_list = []
    age_list = []
    DOB_list = []
    for x in range(df.shape[0]):
        date_time_obj = fake.date_between_dates(date_start=datetime.date(1904, 1, 1), date_end=datetime.date(2019, 12, 31))
        ############
        day = date_time_obj.day
        month = date_time_obj.month
        year = date_time_obj.year
        age = calculate_age_on_31_12_2019(date_time_obj)
        DOB = date_time_obj.strftime("%d%m%Y")
        #######
        day_list.append(day)
        month_list.append(month)
        year_list.append(year)
        age_list.append(age)
        DOB_list.append(DOB)
        #######
    df['Resident_Day_Of_Birth'] = day_list
    df['Resident_Month_Of_Birth'] = month_list
    df['Resident_Year_Of_Birth'] = year_list
    df['Resident_Age'] = age_list
    df['full_DOB'] = DOB_list
```

```python
generate_date_of_birth(person_index_df)
```
